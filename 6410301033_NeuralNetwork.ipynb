```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ตัวอย่างข้อมูล (แก้ไขให้ตรงกับข้อมูลของคุณ)
x = np.linspace(0, 100, 100)  # ข้อมูลแกน x
original = np.sin(2 * np.pi * x / 25)  # ข้อมูลต้นฉบับ (Original)
predicted = original + np.random.normal(0, 0.2, size=len(original))  # ตัวอย่าง Predicted Data

# สร้างฟังก์ชันสำหรับ Plot เปรียบเทียบ
def plot_comparison(x, original, predicted):
    plt.figure(figsize=(10, 6))
    plt.plot(x, original, label="Original", color='blue', linestyle='-')  # เส้น Original
    plt.plot(x, predicted, label="Predict", color='red', linestyle='--')  # เส้น Predicted
    plt.axvline(x=50, color='purple', linestyle='-', label="Split Point")  # เส้นแบ่งข้อมูล
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Comparison between Original and Predicted Data")
    plt.legend()
    plt.grid(True)
    plt.show()

# เรียกใช้ฟังก์ชันเพื่อแสดงกราฟเบื้องต้น
plot_comparison(x, original, predicted)
```

```python
# ฟังก์ชันคำนวณค่า Loss (Mean Squared Error)
def compute_loss(original, predicted):
    return np.mean((original - predicted) ** 2)

# ทดลองปรับพารามิเตอร์ (เช่นการลด Noise)
def optimize_parameters(x, original, noise_range):
    best_noise = 0
    min_loss = float('inf')
    best_predicted = None

    for noise_level in noise_range:
        predicted = original + np.random.normal(0, noise_level, size=len(original))  # ปรับ Noise
        loss = compute_loss(original, predicted)

        if loss < min_loss:
            min_loss = loss
            best_noise = noise_level
            best_predicted = predicted

    return best_noise, best_predicted
```

```python
# ทดลองปรับ Noise
noise_range = np.linspace(0, 0.5, 50)  # ช่วงของ Noise
best_noise, best_predicted = optimize_parameters(x, original, noise_range)
```

```python
plot_comparison(x, original, best_predicted)
```
